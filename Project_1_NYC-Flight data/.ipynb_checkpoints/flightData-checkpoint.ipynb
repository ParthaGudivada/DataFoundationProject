# Domain: Airlines

## Project 01: Analyze NYC-Flight data

    This dataset contains information about all flights that departed from NYC (e.g. EWR, JFK and LGA) 
    in 2013: 336,776 flights in total. 

|Name          |Description  |
|------------------|-------------|
|year |2013|
|month | 1-12 |
|day | Day of the month (1-31) |
|dep_time | Departure times, local timezone |
|sched_dep_time| Scheduled departure time
|dep_delay |Departure delay, in minutes, Negative times represent early departures
   |    arr_time |Arrival times, local timezone
    |sched_arr-time |Scheduled departure time
    |arr_delay | Arrival delay, in minutes, Negative times represent early arrivals
    |carrier| Two letter carrier abbreviation
    |flight | Flight number
    |tailnum |Plane tail number
    |origin, dest | Airport codes for origin and destination
    |air_time| Amount of time spent in the air, in minutes.
    |distance| Distance flown, in miles.
    |hour, minute| Time of departure broken in to hour and mins.
    |time_hour |Timestamp

The following are the types of question you can ask:

### Exploration ideas:
    Departure delays
    Best airports in terms of time departure %.
    Aircraft speed analysis.
    On time arrival % analysis.
    Maximum number of flights headed to some particular destination. 



```python
# Import the relevant modules

import os 
import sys

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl

% matplotlib inline
```

```python
# Read the data file and create the data frame

current_directory = os.getcwd()

file_name = "flight_data.csv"

try:
    df = pd.read_csv(current_directory + '/' + file_name)
    
except FileNotFoundError as e:
    print(file_name + " is not found in " + current_directory)
    
except:
    print("Unexpected error:", sys.exc_info()[0])
    raise

```

```python
# Get the basic info of the data set

df.shape # Check the structure of the data
df.dtypes # check the types of the data
df.isnull().sum() # finding null values
df.columns # printing index
df.dest.unique() 
df.head(10)
```

```python
#  How many flights were there from NYC airports to Seattle ?
len(df[(df.dest=='SEA') & ((df.origin == 'EWR') | (df.origin == 'LGA') | (df.origin == 'JFK'))])

df = pd.DataFrame(df)
#  How many airlines fly from NYC to Seattle?
len(df[(df.dest=='SEA') & ((df.origin == 'EWR') | (df.origin == 'LGA') | (df.origin == 'JFK'))].carrier.unique())

#  How many unique flights fly from NYC to Seattle?
len(df[(df.dest=='SEA') & ((df.origin == 'EWR') | (df.origin == 'LGA') | (df.origin == 'JFK'))].tailnum.unique()) 

#  What is the average arrival delay for flights from NC to Seattle?
df_filtered = df[(df.arr_delay>0) & (df.dest=='SEA') & ((df.origin == 'EWR') | (df.origin == 'LGA') | (df.origin == 'JFK'))]     
print("Average Arrival Delay for Flights from NC To Seattle: ", np.mean(df_filtered.arr_delay))

#  What proportion of flights to Seattle come from each NYC airport?
df_filtered = df[(df.dest=='SEA') & ((df.origin == 'EWR') | (df.origin == 'LGA') | (df.origin == 'JFK'))]     
print("Proportion of Flights to Seatttle coming from NYC airport", np.mean(df_filtered.arr_delay))
```

```python
flights_sea_df = df[(df.dest == 'SEA')]
grouped_data = flights_sea_df.groupby([flights_sea_df.origin]).size()
print (pd.DataFrame({'Percentage':grouped_data / len(flights_sea_df)}))
```

```python
#  What is the largest average departure delay and largest average arrival delay in minutes?
grouped_data = df.groupby([df.month,df.day]).mean()
print(max(grouped_data.arr_delay))
print(max(grouped_data.dep_delay))
```

```python
#  What was the worst month to fly out of NYC?

filtered_df = df[df.dep_delay > 0]

#Grouping by month to find the count of delays
grouped_data = pd.DataFrame(filtered_df.groupby([filtered_df.month])['dep_delay'].size().reset_index(name='Count'))

#Plotting Number of delays per Month
plt.bar(grouped_data.month, grouped_data.Count, color="purple")
plt.ylabel('Number of Delays')
plt.xlabel('Months')
plt.title('Number of delays per Month')
plt.show()

```

```python
# bar chart method
#  On average, how do departure delays vary over the course of a day?

filtered_df = df[df.dep_delay > 0]

#Grouping by hour to find the count of delays
grouped_data = pd.DataFrame(filtered_df.groupby([filtered_df.hour])['dep_delay'].size().reset_index(name='Count'))

#Grouping by hour to find the total count of flights over a day to further find the percentage of delays
grouped_data_non_filtered = pd.DataFrame(df.groupby([df.hour])['dep_delay'].size().reset_index(name='Total_Count'))

#Merging both the dataframes
df_merged = pd.merge(grouped_data, grouped_data_non_filtered, on=['hour'])

df_merged['Percentage_Delays'] = (df_merged.Count/df_merged.Total_Count)*100
print(df_merged)
```

```python

#Plotting Number of delays per Hour of a day

plt.bar(df_merged.hour, df_merged.Percentage_Delays, color="cyan")
plt.ylabel('Percentage of Delays')
plt.xlabel('Hours')
plt.title('Percentage of delays per Hour of a day')
plt.show()
```

```python

# Which flight departing NYC  flew the fastest?
df['speed'] = df['distance']/df['air_time']
print(df[(df.speed == max(df.speed))]['flight'])
```

```python
# What are flights carrier, flight and dest ? Where do they fly to?
grouped_data = pd.DataFrame(df.groupby([df.carrier,df.flight,df.dest]).size().reset_index(name='Count'))
grouped_data[grouped_data.Count == max(grouped_data.Count)]
```

```python
#  Which carriers have been the top and the bottom performers?

#Fitering dataset for flights having arr_delay>0 and dep_delay>0
df_filtered_carrier = df[(df.arr_delay>0) & (df.dep_delay>0)]  

#Grouping by carrier and getting the count
grouped_career_filtered = df_filtered_carrier.groupby([df_filtered_carrier.carrier]).size().reset_index(name='Size')
grouped_career_filtered

#Removing NA's from dep_time column
df_filtered_total = df[np.isfinite(df['dep_time'])] 

#Grouping by carrier to find the total count for each carrier
grouped_career_total = df_filtered_total.groupby([df_filtered_total.carrier]).size().reset_index(name='Total_Size')
grouped_career_total

#Merging both the dataframes
df_col_merged = pd.merge(grouped_career_filtered, grouped_career_total, on=['carrier'])

#CalculatingPercentage delays
df_col_merged['Percentage_Delays'] = (df_col_merged.Size/df_col_merged.Total_Size)*100

df_col_merged
```

```python
# Plotting Percent Delay by Carrier

index = np.arange(len(df_col_merged.carrier))
plt.bar(index, df_col_merged.Percentage_Delays, color="red")
plt.ylabel('Delay Percentage')
plt.xlabel('Carriers')
plt.title('Percent Delay by Carrier through 2013')
plt.xticks(index, df_col_merged.carrier)
plt.show()
```

```python
df_mean_arr_delay = df_filtered_carrier.groupby([df_filtered_carrier.carrier])['arr_delay'].mean().reset_index(name='Mean_Arrival_Delay')

# Plotting Average Arrival Delay for each Carrier
index = np.arange(len(df_mean_arr_delay.carrier))
plt.bar(index, df_mean_arr_delay.Mean_Arrival_Delay, color="green")
plt.ylabel('Avg. Arrival Delay (minutes)')
plt.xlabel('Carriers')
plt.title('Average Arrival Delay for each Carrier')
plt.xticks(index, df_col_merged.carrier)
plt.show()
```

```python
# Analysis -

# 1. We observed that it is worst to fly in the month of JULY.
# 2. We observe that carrier FL has the highest delay %, thus making it the least performer among other carriers. Carrier HA has the best performance in terms of delay %.
# 3. We observe that OO and HA have higher arrival delays among other carriers. UA and US carriers perform best when looking from this perspective. 
```
